```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
observation = 1000

x = np.random.uniform(-10,10,size=(observation,1))
z = np.random.uniform(low=-10,high=10,size=(observation,1))

generated_inputs = np.column_stack((x,z))

noise = np.random.uniform(low=-1,high=1,size=(observation,1))

generated_targets = 2 * x - 7*z +5 +noise

np.savez('npzff', inputs = generated_inputs, targets = generated_targets)
```

```python
training_data = np.load('npzff.npz')
```

```python
input_size = 2 #x and z
output_size = 1 #y

model = tf.keras.Sequential([
                            tf.keras.layers.Dense(output_size,
                                                  kernel_initializer = tf.random_uniform_initializer(minval= -0.1,maxval=0.1),
                                                  bias_initializer = tf.random_uniform_initializer(minval= -0.1,maxval=0.1)
                                                 )
                            ])

custom_optimizer = tf.keras.optimizers.SGD(lr=0.02)
model.compile(optimizer= custom_optimizer, loss = 'mean_squared_error')

model.fit(training_data['inputs'],training_data['targets'],epochs=100, verbose=2)
```
