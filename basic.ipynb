```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
observation = 1000

x = np.random.uniform(-10,10,size=(observation,1))
z = np.random.uniform(low=-10,high=10,size=(observation,1))

inputs = np.column_stack((x,z))

print(inputs.shape)
```

```python
noise = np.random.uniform(low=-1,high=1,size=(observation,1))

target = 2 * x - 7*z +5 +noise

target.shape
```

```python
target = target.reshape(observation)
```

```python
target.shape
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot(x,z,target)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('targets')
ax.view_init(azim=100)
plt.show()
target = target.reshape(observation,1)
```

```python
x[2]
```

```python
init_range = 0.1

weight = np.random.uniform(-init_range,init_range,(2,1))

biases = np.random.uniform(-init_range,init_range,1)
print(weight)
print(biases)
```

```python
learning_rate=0.04
```

```python
for i in range(100):
    outputs = np.dot(inputs,weight) + biases
    deltas = outputs - target
    
    loss = np.sum(deltas ** 2)/2/observation
    
    print(loss)
    
    delta_scaled = deltas/observation
    
    weight = weight - learning_rate*np.dot(inputs.T,delta_scaled)
    biases = biases -learning_rate*np.sum(delta_scaled)
```

```python
print(weight,biases)
```

```python
plt.plot(outputs,target)
plt.xlabel('outputs')
plt.ylabel('targets')
plt.show()
```
